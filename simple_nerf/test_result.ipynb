### 原始均匀采样


```python
z_vals = torch.linspace(near, far, N_samples)
# 所有像素共享相同的采样位置
# 例如: [2.0, 2.0625, 2.125, ..., 6.0]
```



### 添加随机扰动后


```python
z_vals = z_vals + torch.rand(list(rays_o.shape[:-1]) + [N_samples])
#                              ^^^^^^^^^^^^^^^^^^^^^^^^
#                              形状: (H, W, N_samples)
```



**关键点**：
- `torch.rand(...)` 生成形状为 `(H, W, N_samples)` 的随机数
- 这意味着**每个像素 (h, w) 的每个采样点 (n) 都有独立的随机扰动**

### 可视化对比



```python
像素1的光线采样:
|--o----|---o--|----o----|--o-----|...

像素2的光线采样:
|---o---|--o----|--o-----|---o----|...

像素3的光线采样:
|-o-----|----o-|---o-----|--o-----|...
```



每个像素的采样点位置都不同！

## 两层随机性

1. **空间随机**：不同像素的采样点位置不同
2. **时间随机**：每次迭代时，同一像素的采样点位置也会改变

这样可以：
- 更好地覆盖整个采样空间
- 避免所有光线都采样相同位置造成的偏差
- 提高渲染质量和训练稳定性
